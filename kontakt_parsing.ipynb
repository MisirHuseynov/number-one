```
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
class parsing:
  def __init__(self):
    self.a={}
  def kontakt(self):
    i=0
    baza={}
    for x in range(3):
      url=str(input('linki daxil edin:'))
      page=req.get(url)
      soup=bs(page.content,'html.parser')
      marka=soup.find(attrs={'class':'title'}).text.strip()
      qiymet=soup.find(attrs={'class':'gridprice gridprice_cash'}).text.strip()
      baza.update({marka:[qiymet]})
      i+=1
      i
      self.a.update({'kontakt home':baza})
    ab=pd.DataFrame(self.a)
    ab.to_csv('file.csv')
    df=pd.read_csv('file.csv')
    display(baza)
    print('********')
    display(self.a)
    display(df)
netice=parsing()
netice.kontakt()
```
